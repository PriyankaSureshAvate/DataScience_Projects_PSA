```python
# Print the table of Features, F-Score and P-values
columns = list(x.columns)

print (" ")
print (" ")
print (" ")

print ("    Features     ", "F-Score    ", "P-Values")
print ("    -----------  ---------    ---------")

for i in range(0, len(columns)):
    f1 = "%4.2f" % f_score[i]
    p1 = "%2.6f" % p_values[i]
    print("    ", columns[i].ljust(12), f1.rjust(8),"    ", p1.rjust(8))
```

```python
columns = list(x.columns)  # Feature names

# Print table header
print("\n\n")  # Extra spacing for clarity
print(f"{'Features':<15} {'F-Score':>10} {'P-Value':>15}")
print("-" * 40)

# Print each feature with formatted values
for i in range(len(columns)):
    f1 = f"{f_score[i]:.2f}"  # Format F-Score with 2 decimals
    p1 = f"{p_values[i]:.6f}"  # Format P-Value with 6 decimals
    print(f"{columns[i]:<15} {f1:>10} {p1:>15}")
```
